```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('heart_disease_dataset.csv')
df.head()
```

```python
df.loc[df['Gender']=='Male','Gender']=1
df.loc[df['Gender']=='Female','Gender']=0
df.head()
```

```python
df = df.replace('Yes', 1)
df = df.replace('No', 1)
df.head()
```

```python
df["Smoking"].value_counts()
```

```python
df["Chest Pain Type"].value_counts()
```

```python
df["Alcohol Intake"].value_counts()
```

```python
df["Stress Level"].value_counts()
```

```python
df = df.replace('Never', 0)
df = df.replace('Current', 1)
df = df.replace('Former', 2)
df = df.replace('Non-anginal Pain', 1)
df = df.replace('Typical Angina', 2)
df = df.replace('Asymptomatic', 0)
df = df.replace('Atypical Angina', 3)
df = df.replace('Heavy', 2)
df = df.replace('Moderate', 1)
```

```python
df.head()
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```python
df['Alcohol Intake'] = df['Alcohol Intake'].fillna(0)
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred)
print("Accuracy = " + str(acc*100)+"%")
```
